### Обучение нейронной сети с помощью алгоритма Adam

### Сравнение Adam и обучаемого оптимизатора на разных этапах обучения

```python
import numpy as np
import matplotlib.pyplot as plt
loss_test = np.load('pytorch-meta-optimizer/losses/test_simple.npy')
plt.plot(loss_test, label="Adam")
plt.title('Test loss')
epoch = (np.arange(20)*25).tolist()
for ep in epoch[::3]:
    loss_test = np.load('pytorch-meta-optimizer/losses/test_{}.npy'.format(ep))
    plt.plot(loss_test, label="Our #{}".format(ep))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.figure()
loss_test = np.load('pytorch-meta-optimizer/losses/train_simple.npy')
plt.plot(loss_test, label="Adam")
plt.title('Train loss')
epoch = (np.arange(20)*25).tolist()
for ep in epoch[::3]:
    loss_test = np.load('pytorch-meta-optimizer/losses/train_{}.npy'.format(ep))
    plt.plot(loss_test, label="Our #{}".format(ep))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
